```python
import numpy as np
import pandas as pd
import mysql.connector
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from sklearn.preprocessing import MinMaxScaler
from credentials import ipCred, usernameCred, passwordCred, databaseCred

# Database connection parameters
db_config = {
    'host': ipCred,
    'user': usernameCred,
    'password': passwordCred,
    'database': databaseCred
}

# Get ticker from user
ticker = input("Enter the ticker symbol (e.g., AAPL, ORCL): ").strip().upper()

# Establish connection
conn = mysql.connector.connect(**db_config)
query = f"""
    SELECT date, open, high, low, close, volume, total_sentiment
    FROM {ticker}_view
    WHERE date BETWEEN '2024-03-14' AND '2025-3-13'
    ORDER BY date ASC;
"""
df = pd.read_sql(query, conn)
conn.close()

# Convert date column
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)

# Check for NaNs and fill missing values
if df.isnull().values.any():
    print("Warning: NaN values detected in dataset! Filling missing values...")
    df.fillna(method='ffill', inplace=True)  # Forward-fill missing data

# Selecting features
features = ['open', 'high', 'low', 'close', 'volume', 'total_sentiment']
scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df[features])

# Function to create sequences for LSTM
def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length])
        y.append(data[i+seq_length, 3])  # Predicting 'close' price
    return np.array(X), np.array(y)

# Reduce sequence length for better learning
SEQ_LENGTH = 30  # Reduced from 60
X, y = create_sequences(df_scaled, SEQ_LENGTH)

# Train-test split (90% train, 10% test)
split = int(0.9 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# Define LSTM Model with gradient clipping
model = Sequential([
    LSTM(100, return_sequences=True, input_shape=(SEQ_LENGTH, len(features))),
    Dropout(0.2),
    LSTM(100, return_sequences=False),
    Dropout(0.2),
    Dense(50, activation='relu'),
    Dense(1)
])

# Use Adam optimizer with gradient clipping
optimizer = Adam(learning_rate=0.001, clipnorm=1.0)
model.compile(optimizer=optimizer, loss='mae')

# Train the model with increased epochs & reduced batch size
model.fit(X_train, y_train, epochs=200, batch_size=8, validation_data=(X_test, y_test))

# Predictions
y_pred = model.predict(X_test)

# Function to denormalize values
def denormalize(value, scaler, column_index):
    base = np.zeros((value.shape[0], len(features)))  # Create placeholder
    base[:, column_index] = value.flatten()  # Assign predicted values
    return scaler.inverse_transform(base)[:, column_index]

# Denormalize predictions
y_test_actual = denormalize(y_test, scaler, 3)
y_pred_actual = denormalize(y_pred, scaler, 3)

# Debugging check for NaNs
if np.isnan(y_pred_actual).any():
    print("Error: Predicted values contain NaNs!")

# Plot results
plt.figure(figsize=(12,6))
plt.plot(df.index[split + SEQ_LENGTH:], y_test_actual, label="Actual Closing Price", color='blue')
plt.plot(df.index[split + SEQ_LENGTH:], y_pred_actual, label="Predicted Closing Price", color='red')
plt.title(f"{ticker} Stock Price Prediction")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()

# Print first 10 predictions for verification
print("First 10 Predicted Closing Prices:", y_pred_actual[:10])
```

```python
import numpy as np
import pandas as pd
import mysql.connector
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, Bidirectional
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.preprocessing import MinMaxScaler
from credentials import ipCred, usernameCred, passwordCred, databaseCred

# ----------------- Database Connection Parameters -----------------
db_config = {
    'host': ipCred,
    'user': usernameCred,
    'password': passwordCred,
    'database': databaseCred
}

# Get ticker from user
ticker = input("Enter the ticker symbol (e.g., AAPL, ORCL): ").strip().upper()

# ----------------- Load Data from SQL -----------------
conn = mysql.connector.connect(**db_config)
query = f"""
    SELECT date, open, high, low, close, volume, total_sentiment
    FROM {ticker}_view
    WHERE date BETWEEN '2024-03-14' AND '2025-03-13'
    ORDER BY date ASC;
"""
df = pd.read_sql(query, conn)
conn.close()

# Convert date column and set index
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)

# Fill missing values
if df.isnull().values.any():
    print("Warning: NaN values detected in dataset! Filling missing values...")
    df.fillna(method='ffill', inplace=True)

# ----------------- Feature Engineering -----------------
# Optionally, you can add more features; here we add 'range' (high - low)
df['range'] = df['high'] - df['low']

# ----------------- Split Data into Training and Test Sets -----------------
# Train: up to 2025-02-13, Test: 2025-02-14 to 2025-03-13
train_end_date = pd.Timestamp('2025-02-13')
df_train = df[df.index <= train_end_date]
df_test = df[df.index > train_end_date]

print(f"Training data from {df_train.index.min().date()} to {df_train.index.max().date()}")
print(f"Test data from {df_test.index.min().date()} to {df_test.index.max().date()}")

# ----------------- Feature Scaling -----------------
features = ['open', 'high', 'low', 'close', 'volume', 'total_sentiment', 'range']
scaler = MinMaxScaler()
df_train_scaled = scaler.fit_transform(df_train[features])
df_test_scaled = scaler.transform(df_test[features])

# ----------------- Create Sequences from Training Data -----------------
def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length])
        y.append(data[i+seq_length, 3])  # target: 'close' price (index 3)
    return np.array(X), np.array(y)

SEQ_LENGTH = 30
X_train, y_train = create_sequences(df_train_scaled, SEQ_LENGTH)

# ----------------- Build the LSTM Model -----------------
model = Sequential([
    Bidirectional(LSTM(100, return_sequences=True, input_shape=(SEQ_LENGTH, len(features)))),
    Dropout(0.2),
    Bidirectional(LSTM(100, return_sequences=False)),
    Dropout(0.2),
    Dense(50, activation='relu'),
    Dense(1)
])

optimizer = Adam(learning_rate=0.001, clipnorm=1.0)
model.compile(optimizer=optimizer, loss='mae')

early_stop = EarlyStopping(monitor='loss', patience=10, restore_best_weights=True)

# ----------------- Train the Model -----------------
model.fit(X_train, y_train, epochs=200, batch_size=8, validation_split=0.1, callbacks=[early_stop])

# ----------------- Rolling Forecasting (Walk-forward) -----------------
# Use last SEQ_LENGTH days from training as the initial input sequence
current_sequence = df_train_scaled[-SEQ_LENGTH:].copy()
forecasted_scaled = []  # to store predicted scaled 'close' prices

# We'll perform one-step-ahead forecasting using actual test values in a rolling fashion
for i in range(len(df_test_scaled)):
    current_sequence_input = np.expand_dims(current_sequence, axis=0)  # shape (1, SEQ_LENGTH, features)
    next_pred_scaled = model.predict(current_sequence_input)
    forecasted_scaled.append(next_pred_scaled[0, 0])
    # Update the sequence with the actual test data row at position i (rolling window)
    current_sequence = np.vstack((current_sequence[1:], df_test_scaled[i]))

forecasted_scaled = np.array(forecasted_scaled)

# ----------------- Denormalize Predictions -----------------
def denormalize(value, scaler, column_index):
    base = np.zeros((value.shape[0], len(features)))
    base[:, column_index] = value.flatten()
    return scaler.inverse_transform(base)[:, column_index]

forecasted_prices = denormalize(forecasted_scaled, scaler, 3)  # column 3 is 'close'
actual_prices = df_test['close'].values  # actual closing prices

# ----------------- Calculate Accuracy Metrics -----------------
mape = np.mean(np.abs((actual_prices - forecasted_prices) / actual_prices)) * 100
accuracy = 100 - mape

print(f"Mean Absolute Percentage Error (MAPE): {mape:.2f}%")
print(f"Accuracy (100 - MAPE): {accuracy:.2f}%")

# ----------------- Plot the Forecast vs. Actual -----------------
test_dates = df_test.index
plt.figure(figsize=(12,6))
plt.plot(test_dates, actual_prices, label="Actual Closing Price", color='blue')
plt.plot(test_dates, forecasted_prices, label="Forecasted Closing Price", color='red')
plt.title(f"{ticker} Stock Price Forecast")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()

print("First 10 Forecasted Closing Prices:", forecasted_prices[:10])
```
